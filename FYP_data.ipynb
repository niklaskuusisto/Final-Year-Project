# Pre-processing raw data

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy

from scipy import stats
```

```python
# file input and output paths

input_path = 'Data pre-processed excel/'
output_path = 'Data pre-processed python/'
```

```python
# Equities

fig, axes = plt.subplots(3,1) # plot 3 by 1
tickers = ['SPX','DAX','NKY'] #

i=0

for ticker in tickers:
    
    input_file = input_path + ticker + '.csv'
    output_file = output_path + ticker + '.csv'
    
    data = pd.read_csv(input_file, header=0, index_col=0)
    data = data.astype(float) # set to float type
    data.to_csv(output_file) # save to csv
    
    print('\n', ticker)
    print(data.shape)
    print(data.head())
    
    ax = axes[i]
    ax.set_title(ticker)
    ax.plot(data['Close'])
    ax.set_ylabel('Price')
    ax.xaxis.set_major_locator(plt.MaxNLocator(17))
    ax.tick_params(axis='x', labelrotation=45)
    
    i = i+1
    
fig.set_size_inches(12,9)
fig.subplots_adjust(hspace=1)
```
